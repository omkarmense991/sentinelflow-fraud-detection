```python
plt.figure(figsize=(6,4))
sns.countplot(x=df["Class"])
plt.title("Fraud vs Legitimate Transactions")
plt.xlabel("Class")
plt.ylabel("Count")

plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df["Amount"], bins=50)
plt.title("Transaction Amount Distribution")
plt.xlabel("Amount")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(np.log1p(df["Amount"]), bins=50)
plt.title("Transaction Amount Distribution")
plt.xlabel("Amount")
plt.ylabel("Frequency")
plt.show()
```

```python
corr = df.corr()
```

```python
plt.Figure(figsize=(16,12))
sns.heatmap(corr, cmap="coolwarm")
plt.title("Feature Correlation Matrix")

plt.show()
```

## Key Findings

- Dataset is extremely imbalanced (~0.17% fraud)
- Accuracy alone will be misleading
- Transaction amounts are heavily skewed
- Some PCA features show correlation with fraud class
- No missing values exist in the dataset
