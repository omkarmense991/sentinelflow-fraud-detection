```python
from src.data.load_data import load_dataset

df = load_dataset()

fraud_df = df[df["Class"] == 1]
legitimate_df = df[df["Class"] == 0]
print(fraud_df)
print(legitimate_df)
```

```python
fraud_sample_1 = (
    fraud_df
    .iloc[0]
    .drop("Class")
    .to_dict()
)

fraud_sample_2 = (
    fraud_df
    .iloc[1]
    .drop("Class")
    .to_dict()
)

legitimate_sample_1 = (
    legitimate_df
    .iloc[0]
    .drop("Class")
    .to_dict()
)

legitimate_sample_2 = (
    legitimate_df
    .iloc[1]
    .drop("Class")
    .to_dict()
)
```

```python
def convert_to_python_types(sample):

    return {

        key: float(value)

        for key, value in sample.items()
    }


fraud_sample_1 = (
    convert_to_python_types(
        fraud_sample_1
    )
)

fraud_sample_2 = (
    convert_to_python_types(
        fraud_sample_2
    )
)

legitimate_sample_1 = (
    convert_to_python_types(
        legitimate_sample_1
    )
)

legitimate_sample_2 = (
    convert_to_python_types(
        legitimate_sample_2
    )
)
```

```python
import json
import requests

batch_payload = {

    "transactions": [

        fraud_sample_1,

        fraud_sample_2,

        legitimate_sample_1,

        legitimate_sample_2
    ]
}


# =========================================
# Pretty Print Payload
# =========================================

print(
    json.dumps(
        batch_payload,
        indent=2
    )
)
```
